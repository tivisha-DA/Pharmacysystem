```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

from google.colab import drive
drive.mount('/content/drive')
df=pd.read_csv('/content/drive/MyDrive/diabetes.csv')

df.head()

df.head()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Glucose'], bins=20, kde=True)
plt.title('Distribution of Glucose Levels')
plt.xlabel('Glucose')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Outcome', y='BMI', data=df)
plt.title('BMI Distribution by Diabetes Outcome')
plt.xlabel('Diabetes Outcome')
plt.ylabel('BMI')
plt.xticks([0, 1], ['Non-Diabetic', 'Diabetic'])
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Glucose', y='BMI', hue='Outcome', data=df)
plt.title('Scatter Plot of Glucose vs. BMI')
plt.xlabel('Glucose Level')
plt.ylabel('BMI')
plt.legend(title='Diabetes Outcome')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Outcome', data=df, palette='Set2')
plt.title('Count of Diabetic vs. Non-Diabetic Patients')
plt.xticks([0, 1], ['Non-Diabetic', 'Diabetic'])
plt.xlabel('Outcome')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
outcome_counts = df['Outcome'].value_counts()
plt.pie(outcome_counts, labels=['Non-Diabetic', 'Diabetic'], autopct='%1.1f%%', colors=['lightblue', 'orange'])
plt.title('Proportion of Diabetic vs. Non-Diabetic Patients')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='Outcome', y='BloodPressure', data=df) # Fixed: Removed extra indentation
plt.title('Blood Pressure Distribution by Diabetes Outcome')
plt.xlabel('Diabetes Outcome')
plt.ylabel('Blood Pressure')
plt.xticks([0, 1], ['Non-Diabetic', 'Diabetic'])
plt.show()
```
